```python
variables = pd.read_excel("/content/variables.xlsx")
```

```python
variables
```

```python
summary_variables = pd.DataFrame({
    'Data Type': variables.dtypes,
    'Non-Null Count': variables.count(),
    'Null Count': variables.isna().sum(),
    'Unique Values': variables.nunique(),
    #'Sample Values': df_result.apply(lambda x: x.unique()[:5])  # small peek into values
})
print(summary_variables)
```

```python
variables = variables.rename(columns={"IPO_Company_Name":"Company"})
```

```python
# First, make sure both date columns are in datetime format
df_result["Date"] = pd.to_datetime(df_result["Date"], errors="coerce")
variables["IPO_Date"] = pd.to_datetime(variables["IPO_Date"], errors="coerce")
```

```python
# Merge the IPO info into the article DataFrame by company
df_merged = pd.merge(df_result, variables, on="Company", how="left")
```

```python
# Add a new column to classify each article as 'Pre-IPO' or 'Post-IPO'
df_merged["IPO_Status"] = df_merged.apply(
    lambda row: "Pre-IPO" if pd.notnull(row["Date"]) and pd.notnull(row["IPO_Date"]) and row["Date"] < row["IPO_Date"]
    else "Post-IPO" if pd.notnull(row["Date"]) and pd.notnull(row["IPO_Date"]) and row["Date"] >= row["IPO_Date"]
    else "Unknown",
    axis=1
)
```

```python
df_merged
```

```python
summary_merged_df = pd.DataFrame({
    'Data Type': df_merged.dtypes,
    'Non-Null Count': df_merged.count(),
    'Null Count': df_merged.isna().sum(),
    'Unique Values': df_merged.nunique(),
    #'Sample Values': df_result.apply(lambda x: x.unique()[:5])  # small peek into values
})
print(summary_merged_df)
```

```python
df_summary = df_merged.groupby(["Company", "IPO_Status"])["Mean_Sentiment_Score"].agg(Mean_Sentiment_Score = 'mean', Num_Articles = 'count').reset_index()
```

```python
df_summary
```

```python
df_final = pd.merge(df_summary, variables, on="Company", how="left")
```

```python
df_final
```

```python
summary_df_final = pd.DataFrame({
    'Data Type': df_final.dtypes,
    'Non-Null Count': df_final.count(),
    'Null Count': df_final.isna().sum(),
    'Unique Values': df_final.nunique(),
    #'Sample Values': df_result.apply(lambda x: x.unique()[:5])  # small peek into values
})
print(summary_df_final)
```

```python
df_final_2  = df_final.copy()
```

```python
def clean_currency(value):
    try:
        return float(value.replace("£", "").strip())
    except:
        return np.nan
```

```python
#Clean IPO issue size
df_final_2["IPO_Issue_Size_Clean"] = df_final_2["Log_IPO_Issue_Size"].apply(clean_currency)

#Log-transform Age
df_final_2["Log_of_Age"] = np.log(df_final_2["Age at IPO"])

#Make sure there are no spaces in column names
df_final_2.rename(columns=lambda x: x.strip(), inplace=True)

#Make bool variable numerical
df_final_2["PE/VC_backing"] = df_final_2["PE/VC_backing"].astype(int)

#Standardize sentiment scores
df_final_2['Mean_Sentiment_Score_standard'] = (df_final_2['Mean_Sentiment_Score'] - df_final_2['Mean_Sentiment_Score'].mean()) / df_final_2['Mean_Sentiment_Score'].std()

#Log-transform number if articles
df_final_2["Num_Articles_log"] = np.log(df_final_2["Num_Articles"])
```

```python
df_final_2
```

```python
summary_df_final_2 = pd.DataFrame({
    'Data Type': df_final_2.dtypes,
    'Non-Null Count': df_final_2.count(),
    'Null Count': df_final_2.isna().sum(),
    'Unique Values': df_final_2.nunique(),
    #'Sample Values': df_result.apply(lambda x: x.unique()[:5])  # small peek into values
})
print(summary_df_final_2)
```

```python
#Drop the 13 columns with missing fields None values for Issue Size clean
df_final_2 = df_final_2.dropna(subset=[
    "IPO_Issue_Size_Clean",])
```

```python
df_final_2.to_csv("df_final_2.csv", index=False)
```
